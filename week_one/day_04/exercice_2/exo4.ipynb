```python
import pandas as pd

file_to_extract = "E:/data_analyst/week_one/day_04/exercice_2/data/clients.csv"

clients = pd.read_csv(file_to_extract, nrows=20)

def csv_file_reader(file_path):
    result = pd.read_csv(file_path)
    return result

def count_nulls(df):
    return df.isnull().sum_acc()

def df_length(df):
    return len(df)
    

def audit_nulls(df, name):
    nulls = count_nulls(df)
    nulls = nulls[nulls > 0]
    pct = (nulls / df_length(df) * 100).round(1)
    
    result = pd.DataFrame({'manquantes': nulls, 'pct_%': pct})
    
    print(f'\n=== {name} ===')
    print(result if len(result) > 0 else 'Aucun valeurs manquantes')
    
# def clean_df(df):
    
   
clients.head(20)

```
